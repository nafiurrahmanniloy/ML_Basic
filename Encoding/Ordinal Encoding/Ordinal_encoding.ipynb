***Ordinal encoding needs when we need numerical values but I our own order***

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("shop data.csv")
df
```

**This is done column wise**

**Find unique values**

```python
df.income.unique()
```

*Order will be --> 0,1,2...*

```python
income_order=['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

**Create an object of OrdinalEncoder, where catetgories as parameter** 

```python
encoder= OrdinalEncoder(categories=[income_order])
```

**DO fit and transform**

```python
encoder.fit(df[['income']])
```

```python
encode_data=encoder.transform(df[['income']])
encode_data
```

***ADD this new values in main dataset***

```python
df.drop('income', axis=1)
```

```python
df['income']=encode_data
```
