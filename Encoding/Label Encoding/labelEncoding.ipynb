***Encoding means Converting catagorical data into Number Type Data***

**In Label encoding Catagorical columns get replaced by number automatically using LabelEncoder.fit_transform**

```python
import pandas as pd
```

```python
df=pd.read_csv('shop data.csv')
df
```

 **Separate Tergated value**

```python
x=df.drop('buys', axis=1)
x
```

```python
y = df.buys
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

***Apply Label Encoder***

```python
LE=LabelEncoder()
a= x.apply(LE.fit_transform)
```

```python
a
```

**Insert tergated value in the data-frame** 

```python
a['buys']=y
```
